```python
import cx_Oracle as ora
def insert(ary):
    for row in ary:
        connections = ora.connect(oracle)
        cursor = connections.cursor()
        sql_insert = 'insert into  welfarefacilities VALUES((select nvl(max(facilityNo),0)+1 from welfarefacilities),\
        :type, :name, :loc, :address,:tell, :x, :y, :score)'
        try:
            cursor.execute(sql_insert,type=row[0].split("(")[0],name=row[1], loc = row[2].split(" ")[0],
                           address = row[2],tell=row[3],x = row[6],y = row[7],score = np.random.randint(100))
            connections.commit()
        except Exception as e:
            print(e)
        finally:
            connections.close()
insert(ary)
```
